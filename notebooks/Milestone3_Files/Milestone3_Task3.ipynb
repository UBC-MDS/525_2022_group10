### 6. Preprocessing

As we can see there is huge range of values in each of these models. Hence we need to normalize them. To do so we are using the `StandardScaler()` trnaformation from Sklearn

```python
numeric_transformer = make_pipeline(StandardScaler())
```

```python
numeric_features = list(X_train.columns.values)
```

```python
preprocessor = make_column_transformer((numeric_transformer, numeric_features),)
```

### 6. Model Building

```python
model = make_pipeline(preprocessor, RandomForestRegressor())
model.fit(X_train, y_train)
print(f"Train RMSE: {mean_squared_error(y_train, model.predict(X_train), squared=False):.2f}")
print(f"Test RMSE: {mean_squared_error(y_test, model.predict(X_test), squared=False):.2f}")
```
